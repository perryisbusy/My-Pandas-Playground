# The left_on and right_on Parameters

```python
week1 = pd.read_csv("/Volumes/GoogleDrive/My Drive/Udemy/Pandas/Data Source/pandas/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("/Volumes/GoogleDrive/My Drive/Udemy/Pandas/Data Source/pandas/Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("/Volumes/GoogleDrive/My Drive/Udemy/Pandas/Data Source/pandas/Restaurant - Customers.csv")
foods = pd.read_csv("/Volumes/GoogleDrive/My Drive/Udemy/Pandas/Data Source/pandas/Restaurant - Foods.csv")
```

# Merging by Index with left_index and right_index Parameters

```python
week1 = pd.read_csv("/Volumes/GoogleDrive/My Drive/Udemy/Pandas/Data Source/pandas/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("/Volumes/GoogleDrive/My Drive/Udemy/Pandas/Data Source/pandas/Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("/Volumes/GoogleDrive/My Drive/Udemy/Pandas/Data Source/pandas/Restaurant - Customers.csv",
                       index_col="ID")
foods = pd.read_csv("/Volumes/GoogleDrive/My Drive/Udemy/Pandas/Data Source/pandas/Restaurant - Foods.csv", 
                   index_col="Food ID")
```

```python
customers.head()
```

```python
foods.head()
```

```python
week1.head()
```

```python
# week1 merge customers into sales
sales = week1.merge(right=customers, how="left", left_on="Customer ID", right_index=True, 
                    sort=False, indicator=True).drop(labels="_merge", axis="columns")
sales.head()
```

```python
# sales merge foods into sales
sales = sales.merge(right=foods, how="left", left_on="Food ID", right_index=True, sort=True, 
                    indicator=True)
sales.head()
```

```python
# merge week1 and week2 using indexes
```

```python
week1.merge(right=week2, how="left", left_index=True, right_index=True, 
            suffixes=["- Week 1", " - Week 2"], sort=True, indicator=True)
```
