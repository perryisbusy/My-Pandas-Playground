# The .sort_index() Method on a MultiIndex DataFrame

```python
big_mac_location = "/Volumes/GoogleDrive/My Drive/LIUZHICHENG/Udemy/Pandas/Data Source/pandas/bigmac.csv"
bigmac= pd.read_csv(filepath_or_buffer=big_mac_location, parse_dates=["Date"],  
                   index_col=["Date", "Country"], squeeze=False)
bigmac.sort_index(ascending=True, inplace=True)
bigmac.head()
```

```python
# two indexes will be sorted by ascending order
bigmac.sort_index()

bigmac.sort_index(ascending=True, inplace=False)
```

```python
# sort Date -> ascending and Country -> descending
bigmac.sort_index(axis=0, ascending=[True, False], inplace=False)
```

```python
# sort Date -> ascending and Country keeps the existing order
bigmac.sort_index(axis=0, level=0, ascending=True, inplace=False)

bigmac.sort_index(axis=0, level="Date", ascending=True, inplace=False)
```

```python
# sort Country -> ascending and keep the Date as the original order
bigmac.sort_index(axis=0, level=1, ascending=False, inplace=False)

bigmac.sort_index(axis=0, level="Country", ascending=False, inplace=False)
```

# The swaplevel(i=, j=, axis=) Method

```python
big_mac_location = "D:/LIUZHICHENG/Udemy/Pandas/Data Source/pandas/bigmac.csv"
bigmac= pd.read_csv(filepath_or_buffer=big_mac_location, parse_dates=["Date"],  
                   index_col=["Date", "Country"], squeeze=False)
bigmac.sort_index(ascending=True, inplace=True)
bigmac.head()
```

```python
# swap the multi index level 
bigmac.swaplevel()
```

```python
# the order doesn't matter since the the multi index only has two levels
# swaplevel() doesn't have the inplace parameter, so you have to replace manually
bigmac.swaplevel("Date", "Country")

bigmac.swaplevel("Country", "Date")

bigmac.swaplevel(0, 1)

bigmac.swaplevel(1, 0)
```

```python
bigmac.swaplevel()
```

```python
# swaplevel() with more than three indexes
bigmac.reset_index(drop=False, inplace=True)
bigmac.set_index(keys=["Date", "Country", "Price in US Dollars"], inplace=True)
bigmac.head()
```

```python
# swap the "Date" column and "Price in US Dollars" column
bigmac.swaplevel(i=0, j=2, axis=0)
```

```python
# swap the "Country" column and "Price in US Dollars" column
bigmac.swaplevel(i=1, j=2, axis=0)
```

# The .unstack() Method, Part 1

```python
worldstats_location = "D:/LIUZHICHENG/Udemy/Pandas/Data Source/pandas/worldstats.csv"
worldstats = pd.read_csv(filepath_or_buffer=worldstats_location, index_col=["country", "year"])
worldstats.head()
```

```python
s = worldstats.stack()
s
```

```python
# return to the original dataframe
s.unstack()
```

```python
# unstack the year index, 列 year -> 行 year
s.unstack().unstack()
```

```python
# unstack the population
s.unstack().unstack().unstack()
```

# The .unstack() Method, Part 2

```python
s = worldstats.stack()
s
```

```python
# country -> index 0
# year -> index 1
# population, GDP -> index 2
s.unstack(level=0)
```

```python
# shift year as a column index
s.unstack(level=1)
```

```python
# 
s.unstack(level=-1)
# same
s.unstack(level=2)
```

```python
s.unstack(level=-2)
# same
s.unstack(level=1)
# 
s.unstack(level="year")
```

# The .unstack() Method, Part 3

```python
# countty -> index 0
# year -> index 1
s
```

```python
# country first then year
s.unstack(level=[0, 1])
```

```python
# year first then country
s.unstack(level=[1, 0])
```

```python
s.unstack(level=["country", "year"])
```

```python
s.unstack(level="year", fill_value=0)
```

# Conclusion:
# df.stack(level=) 将数据的列索引转换为行索引, 增加行
# df.unstack(level=, fill_value=) 将数据的行索引转换为列索引, 减少行

# The pivot_table() Method

```python
food_location = "D:/LIUZHICHENG/Udemy/Pandas\\Data Source\\pandas/foods.csv"
foods = pd.read_csv(filepath_or_buffer=food_location)
foods.head()
```

```python
foods.pivot_table(values="Spend", index="Gender", columns="Item", aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], columns="City", aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], columns=["Frequency", "City"], 
                  aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], columns="City", aggfunc="count")
```

```python
pd.pivot_table(data=foods, values="Spend", index=["Gender", "Item"], columns="City", 
               aggfunc="max")
```
