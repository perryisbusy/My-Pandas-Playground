# Rename Index Labels or Columns in a DataFrame

# Filtering with the .where() Method
# However, .where() will retrieve the whole dataframe

```python
bond = pd.read_csv(filepath_or_buffer=bond_location, index_col=["Film"])
bond.sort_index(ascending=True, inplace=True)
bond.head()
```

```python
mask1 = bond["Actor"] == "Sean Connery"
bond[mask1]
```

```python
# return the entire DataFrame
# cond -> condition
bond.where(cond=mask1)
```

```python
bond.where(cond=bond["Box Office"] > 800, inplace=False)
```

```python
mask2 = bond["Box Office"] > 800
```

```python
bond[mask1 & mask2]
```

```python
bond.where(cond = mask1 & mask2)
```

# A Review of the .apply() Method on Columns

```python
bond = pd.read_csv(filepath_or_buffer=bond_location, index_col=["Film"])
bond.sort_index(ascending=True, inplace=True)
bond.head()
```

```python
bond.columns = [column_name.replace(" ", "_") for column_name in bond.columns]
bond.head(1)
```

```python
bond["Year"].apply(func=lambda x:x*2)
```

```python
def conver_to_string_and_add_millions(number):
    return str(number) + " MILLIONS!"
```

```python
bond["Box_Office"].apply(func=conver_to_string_and_add_millions)
```

```python
# bond.loc[:, []]
bond.loc[:,"Budget"].apply(func=conver_to_string_and_add_millions)
```

```python
columns = ["Box_Office", "Budget", "Bond_Actor_Salary"]
for column in columns:
    bond[column] = bond[column].apply(conver_to_string_and_add_millions)
```

```python
bond.head()
```

# The .apply() Method with Row Values

```python
bond.sort_index(ascending=True, inplace=True)
bond.head()
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have not clue"
```

```python
# axis = 1 axis = "columns"
# a little bit confusing since using column not the row
bond.apply(func=good_movie, axis=1)
```
