# The .duplicated() Method

```python
employee_location = "/Volumes/GoogleDrive/My Drive/LIUZHICHENG/Udemy/Pandas/Data Source/pandas/employees.csv"
```

```python
emp = pd.read_csv(filepath_or_buffer=employee_location, 
                  parse_dates=["Start Date", "Last Login Time"])
emp["Team"] = emp["Team"].astype(dtype="category")
emp["Gender"] = emp["Gender"].astype(dtype="category")
emp["Senior Management"] = emp["Senior Management"].astype(dtype=bool)
```

```python
emp.sort_values(by="First Name", ascending=True, inplace=True)
```

```python
emp["First Name"]
```

```python
# - ``first`` : Mark duplicates as ``True`` except for the first occurrence.
# - ``last`` : Mark duplicates as ``True`` except for the last occurrence.
# -   False : Mark all duplicates as ``True``.
```

```python
# the first value appears won't be marked as duplicated
# by default keep = "first"
emp["First Name"].duplicated()
```

```python
# keep the first one
# by default keep = "first"
mask1 = emp["First Name"].duplicated(keep="first")
emp[mask1]
```

```python
emp["First Name"].head(15)
```

```python
# it got subset
emp.duplicated(subset=["Gender"], keep="first")
```

```python
# keep the last one
emp["First Name"].duplicated(keep="last").head(15)
```

```python
# all duplicated will be marked True
emp["First Name"].duplicated(keep=False)
```

```python
# False : Mark all duplicates as ``True``.
# only none duplicated will be kept
# rarely used
mask2 = emp["First Name"].duplicated(keep=False)
emp[mask2]
```

```python
# true doesn't work and parameter "keep" only has "first", "last", True
emp["First Name"].duplicated(keep=True)
```

```python
# remove all duplicated value and just keep the unique first name
# tilde ~  波浪号 取反
mask3 = ~emp["First Name"].duplicated(keep=False)
emp[mask3]
```

# The .drop_duplicates() Method

```python
emp = pd.read_csv(filepath_or_buffer=employee_location, 
                 parse_dates=["Start Date", "Last Login Time"])
emp["Gender"] = emp["Gender"].astype(dtype="category")
emp["Senior Management"] = emp["Senior Management"].astype(dtype=bool)
emp["Team"] = emp["Team"].astype("category")
```

```python
emp.info()
```

```python
emp.sort_values(by="First Name", ascending=True, inplace=True)
```

```python
emp.head()
```

```python
len(emp)
```

```python
# it won't change since there are no two identical rows
emp.drop_duplicates()
```

```python
# - ``first`` : Mark duplicates as ``True`` except for the first occurrence.
# - ``last`` : Mark duplicates as ``True`` except for the last occurrence.
# -   False : Mark all duplicates as ``True``.
```

```python
# keep the first value that occurs firstly
emp.drop_duplicates(subset=["First Name"], keep="first")
# keep the last value that occurs lastly
emp.drop_duplicates(subset="First Name", keep="last")
# remove all values occuring more than once
emp.drop_duplicates(subset=["First Name"], keep=False)
```

```python
# every category in Team occurs more than once
emp.drop_duplicates(subset=["Team"], keep=False)
```

```python
# only "First Name" and "Team" are same so that it will remove otherwise keeping it
# 101 and 440
# 137, 141, 302
emp.drop_duplicates(subset=["First Name", "Team"], keep="first", inplace=False)
```

```python
# duplicated() returns the ones that are duplicate without the first row
emp[~emp.duplicated(subset=["Team", "Gender"])]
# drop_duplicates() drop duplicate rows without the first row
emp.drop_duplicates(subset=["Team", "Gender"])
```
