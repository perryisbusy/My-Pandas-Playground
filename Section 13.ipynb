# Creating Bar Graphs to Show Counts

```python
def rank_performance(stock_price):
    if stock_price <= 10:
        return "Poor"
    elif stock_price <= 50:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
plt.style.use(style='dark_background')
bb["Close"].apply(func=rank_performance).value_counts(ascending=True).plot(kind="barh")
```

# Creating Pie Charts to Represent Proportions

```python
bb["Close"].mean()
```

```python
def rank_performance(stock_pirce):
    if stock_pirce >= bb["Close"].mean():
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use(style="seaborn")
bb["Close"].apply(func=rank_performance).value_counts().plot(kind="pie", legend=True, 
                                                             figsize=(15, 10))
```
