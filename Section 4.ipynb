# Rank Values with the .rank() Method

```python
nba = pd.read_csv("C:\\Users\\Perry\\Google Drive\\LIUZHICHENG\\Udemy\\Pandas\\Data Source\\pandas\\nba.csv")
nba.head()
```

```python
nba.dropna(axis=0, how="all", inplace=True)
nba["Salary"].fillna(value=0, inplace=True)
```

```python
# .astype() must use before cleaning all NaN
nba["Salary"] = nba["Salary"].astype(dtype="int")
nba["Number"] = nba["Number"].astype(dtype="int")
nba["Age"] = nba["Age"].astype(dtype="int")
```

```python
nba.info()
```

```python
# return ranking column for every column!
nba.rank(ascending=True)
```

```python
# return ranking column for every column!
nba.rank(ascending=False)
```

```python
# return a ranking column
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype(dtype="int")
```

```python
nba.sort_values(by=["Salary"], ascending=False)
```

```python
nba.sort_values(by="Salary Rank", ascending=True).head(20)
```
