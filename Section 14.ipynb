```python
data = np.random.randint(low=0, high=100, size=[1000, 50])
data
```

```python
df = pd.DataFrame(np.random.randint(low=0, high=100, size=[1000, 50]))
df
```

```python
df
```

```python
df
```

```python
pd.options.display.max_columns
```

# Changing pandas Options with Methods

```python
data = np.random.randint(low=0, high=100, size=[1000, 50])
df = pd.DataFrame(data)
df.tail()
```

```python
pd.get_option("max_rows")
```

```python
pd.get_option("max_columns")
```

```python
pd.set_option("max_rows", 1)
df
```

```python
pd.set_option("max_columns", 1)
df
```

```python
pd.options.display.max_columns
```

```python
pd.options.display.max_rows
```

```python
pd.reset_option("max_rows")
```

```python
pd.get_option("max_rows")
```

```python
pd.reset_option("max_columns")
```

```python
pd.get_option("max_columns")
```

```python
pd.describe_option("max_columns")
```

# The precision Option
