# The .agg() Method

```python
fortunes_location = "D:\\LIUZHICHENG\\Udemy\\Pandas\\Data Source\\pandas\\fortune1000.csv"
fortunes = pd.read_csv(fortunes_location, index_col=["Rank"])
sectors = fortunes.groupby(by=["Sector"], axis=0)
# group by Sector
```

```python
# 'DataFrameGroupBy' object has no attribute 'loc'
sectors.loc[:, "Revenue"].sum()
```

```python
sectors["Revenue"].sum().to_frame()
```

```python
# same as the above one
sectors.agg({"Revenue":"sum"})
```

```python
# based on the sectors groups,
# return sum of the Revenue, median of the Profits and mean of the Employees
sectors.agg(func={"Revenue":"sum", "Profits":"median", "Employees":"mean"})
```

```python
# return size, sum and mean for every numerical columns
sectors.agg(func=["size", "sum", "mean", "median", "max", "min"])
```

```python
# lots of combinations
sectors.agg(func={"Revenue":["sum", "mean"], "Profits":["mean", "median"]})
```

# Iterating through Groups

```python
fortunes_location = "D:\\LIUZHICHENG\\Udemy\\Pandas\\Data Source\\pandas\\fortune1000.csv"
fortunes = pd.read_csv(fortunes_location, index_col=["Rank"])
sectors = fortunes.groupby(by=["Sector"], axis=0)
# group by Sector
```

```python
df = pd.DataFrame(columns=fortunes.columns)
df
```

```python
# sector as the group object and data as rest of data
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(n=1, columns="Revenue")
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

```python
cities = fortunes.groupby(by=["Location"])
df = pd.DataFrame(columns=fortunes.columns)
df
```

```python
for city, data in cities:
    highest_revenue_city_in_group = data.nlargest(n=1, columns="Revenue")
    df = df.append(highest_revenue_city_in_group)
```
